# Ejemplo: Normal-Normal con varianza conocida

## 1- Original

```python
from scipy.stats import norm
```

```python
# Semilla para reproducibilidad
np.random.seed(2)

# Parámetros
sigma = 2
numdatos = 10
verdadera_media = 5

# Generar datos
dat = np.random.normal(verdadera_media, sigma, numdatos)

#Grafica
r = np.linspace(-7, 10, 1000)

#Hiperparámetros
tau = 0
rho = 2

# Previa
plt.plot(r, norm.pdf(r, tau, rho), label="previa", color="blue", linewidth=2)

# Verosimilitud
ver = np.array([np.exp(-(np.sum(dat - x)**2) / (2 * sigma**2)) for x in r])
c = 0.9
plt.plot(r, c * ver, label="verosimilitud", color="red", linewidth=2)

# Posterior
A = len(dat) / sigma**2 + 1 / rho**2
B = np.sum(dat) / sigma**2 + tau / rho**2
plt.plot(r, norm.pdf(r, B / A, np.sqrt(1 / A)), label="posterior", color="black", linewidth=2)

# Leyenda
plt.legend(loc="upper left", frameon=False)

# Mostrar gráfica
plt.ylim(0, 1)
plt.show()
```
